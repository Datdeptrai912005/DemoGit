```
import pandas as pd
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from tqdm import tqdm
import time

# 1. ĐỌC DỮ LIỆU
file_path = '/content/drive/MyDrive/MLLA /public_train.csv'
df = pd.read_csv(file_path)

print("✅ ĐỌC FILE THÀNH CÔNG!")
print(f"Tổng số mẫu: {len(df)}")
print(f"Phân bố nhãn:\n{df['label'].value_counts()}")

# 2. TẢI MÔ HÌNH
print("\n--- ĐANG TẢI MÔ HÌNH ---")
model_name = "Qwen/Qwen2.5-1.5B-Instruct"
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)
print("✅ TẢI MÔ HÌNH THÀNH CÔNG!")

# 3. CẢI THIỆN PROMPT
def create_better_prompt(text):
    prompt = f"""Hãy xác định bài đăng sau là tin THẬT hay tin GIẢ.

Quy tắc:
- Tin THẬT: thông tin chính xác, có nguồn gốc rõ ràng, được kiểm chứng
- Tin GIẢ: thông tin sai sự thật, tin đồn thất thiệt, thông tin gây hoang mang

Bài đăng: "{text}"

Bài đăng này là (chỉ trả lời THẬT hoặc GIẢ):"""
    return prompt

def predict_with_confidence(text, model, tokenizer, device):
    """Dự đoán với confidence score"""
    prompt = create_better_prompt(text)

    messages = [{"role": "user", "content": prompt}]
    text_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    inputs = tokenizer([text_prompt], return_tensors="pt", truncation=True, max_length=512).to(device)

    with torch.no_grad():
        outputs = model.generate(
            inputs.input_ids,
            max_new_tokens=15,
            temperature=0.3,
            do_sample=True,
            top_p=0.9,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Phân tích response
    response_lower = response.lower()
    if "thật" in response_lower[-50:]:
        return 1, "THẬT"
    elif "giả" in response_lower[-50:]:
        return 0, "GIẢ"
    else:
        return -1, "KHÔNG RÕ"

# 4. CHẠY THỬ VỚI 10 MẪU (có hiển thị tiến độ)
print("\n--- BẮT ĐẦU DỰ ĐOÁN 10 MẪU ---")

results = []
for idx in tqdm(range(10), desc="Đang xử lý"):
    row = df.iloc[idx]
    text = str(row['post_message'])
    true_label = row['label']

    # Bỏ qua URL và text quá ngắn
    if text == '<URL>' or len(text) < 20:
        print(f"\nMẫu {idx+1}: Bỏ qua (URL hoặc text quá ngắn)")
        continue

    pred, pred_label = predict_with_confidence(text, model, tokenizer, device)

    results.append({
        'index': idx,
        'text': text[:100] + '...',
        'true': true_label,
        'true_label': 'GIẢ' if true_label == 0 else 'THẬT',
        'pred': pred,
        'pred_label': pred_label,
        'correct': pred == true_label
    })

    # In kết quả ngay sau khi dự đoán
    print(f"\n--- MẪU {idx+1} ---")
    print(f"Text: {text[:150]}...")
    print(f"Nhãn thật: {true_label} ({'GIẢ' if true_label==0 else 'THẬT'})")
    print(f"Dự đoán: {pred_label}")
    print(f"Kết quả: {'✅ ĐÚNG' if pred == true_label else '❌ SAI'}")

    time.sleep(0.5)  # Tạm dừng để dễ đọc

# 5. TỔNG KẾT
print("\n" + "="*50)
print("KẾT QUẢ TỔNG HỢP:")
print("="*50)

df_results = pd.DataFrame(results)
if len(df_results) > 0:
    accuracy = df_results['correct'].mean()
    print(f"Số mẫu dự đoán: {len(df_results)}")
    print(f"Số mẫu đúng: {df_results['correct'].sum()}")
    print(f"Accuracy: {accuracy:.2%}")

    print("\nChi tiết từng mẫu:")
    for _, row in df_results.iterrows():
        mark = "✅" if row['correct'] else "❌"
        print(f"{mark} Mẫu {row['index']+1}: Thật={row['true_label']}, Dự đoán={row['pred_label']}")
else:
    print("Không có mẫu nào được dự đoán!")

# 6. THỬ VỚI TEMPERATURE THẤP HƠN
print("\n" + "="*50)
print("THỬ VỚI TEMPERATURE=0.1 (CHẮC CHẮN HƠN)")
print("="*50)

def predict_deterministic(text, model, tokenizer, device):
    """Dự đoán với temperature thấp"""
    prompt = create_better_prompt(text)
    messages = [{"role": "user", "content": prompt}]
    text_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    inputs = tokenizer([text_prompt], return_tensors="pt", truncation=True, max_length=512).to(device)

    with torch.no_grad():
        outputs = model.generate(
            inputs.input_ids,
            max_new_tokens=10,
            temperature=0.1,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response_lower = response.lower()

    if "thật" in response_lower[-50:]:
        return 1
    elif "giả" in response_lower[-50:]:
        return 0
    return -1

# Thử lại với mẫu đầu tiên
print("\nThử lại với mẫu 1 (temperature=0.1):")
text1 = str(df.iloc[0]['post_message'])
pred1 = predict_deterministic(text1, model, tokenizer, device)
print(f"Nhãn thật: 0 (GIẢ)")
print(f"Dự đoán: {pred1} ({'THẬT' if pred1==1 else 'GIẢ' if pred1==0 else 'Không rõ'})")
```
